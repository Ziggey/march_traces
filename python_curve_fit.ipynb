```python
from scipy.optimize import curve_fit
from scipy import polyfit
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
%matplotlib inline

'''
This notebook is based on https://www.youtube.com/watch?v=Kk5_3QAEFuI
'''
```

```python
#y = mx + b

n = 50
jitter_amp = 4.0
x = np.linspace(0,10,n)
jitter = jitter_amp * (np.random.random(n)-.5)
y = x + jitter
plt.plot(x,y,'bo')
```

```python
m, b = polyfit(x,y,1)
[m,b]
```

```python
plt.plot(x,y,'o')
t = np.array([0,10])
plt.plot(t, m*t+b)
plt.tight_layout()
```

```python
def fit_func(t, m, b):
    return m*t+b
```

```python
p, cov = curve_fit(fit_func, x, y)
p
```

```python
plt.plot(x,y,'o')
t = np.array([0,10])
plt.plot(t, p[0]*t+p[1])
plt.xlabel('x')
plt.ylabel('y')
plt.title('linear Fit of Nosiy Data')
plt.legend(['Noisy Data', 'Linear Fit to Noisy Data'], loc=0)
```

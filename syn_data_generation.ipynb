## Obtain Autocorrelation

```python
from statsmodels import api as sm
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
from statsmodels.tsa.statespace.tools import diff
```

```python
acfdf = sm.tsa.acf(jitter, nlags=len(jitter))

lag_df  = np.arange(len(jitter))
plt.plot(lag_df, acfdf,'o')
#plt.xlim((0, 2000))
plt.xlabel('Lags')
plt.ylabel('Autocorrelation')
```

```python
def autocrr(x,nlag):
    
    acf = sm.tsa.acf(x, nlags=nlag)
    plt.figure(figsize = (10, 8))
    lag = np.arange(len(x))
    plt.plot(lag, acf)
    plt.xlabel('Lags')
    plt.ylabel('Autocorrelation')
    
    
    
```

```python
autocrr(jitter[:1000],1000)
```

```python
def autocrr2(x):
    
    n = np.arange(50,len(x),100)
    
    for i in n:
        acf = sm.tsa.acf(x[:i], nlags=i)
        plt.figure(figsize = (10, 8))
        lag = np.arange(len(x[:i]))
        
        
        plt.plot(lag, acf)
        
        plt.xlabel('No. of Lags: ' + str(i))
        plt.ylabel('Autocorrelation')
   
```

```python
autocrr2(jitter[:1000])
```

```python
def autocrr_peak(x):
    
    n = np.arange(50,len(x),100)
    
    acfs = []
    peaks_pos = []
    
    for i in n:
        acf = sm.tsa.acf(x[:i], nlags=i)
        plt.figure(figsize = (10, 8))
        lag = np.arange(len(x[:i]))
                        
        peaks_positive, _ = signal.find_peaks(acf, height=0.0, threshold=None)
                   
        plt.plot(lag, acf,'b-')
        plt.plot(lag[peaks_positive], acf[peaks_positive], 'ro')
        print (peaks_positive, 'for : ', i)
        plt.xlabel('No. of Lags: ' + str(i))
        plt.ylabel('Autocorrelation')
        
    
```

```python
xe = np.linspace(0,1, num = 40)
ye = 3.45 * np.exp(-(1.334 * xe)) 

plt.plot(xe,ye,'ro')
```

```python
def test_func(x, a, b):
    return a*np.exp(-(b*x))
```

```python
p, cov = curve_fit(test_func, xe, ye)
p
```

```python
plt.plot(xe, ye, 'o', color ='red', label ="data") 
plt.plot(xe, (p[0]*np.exp(-(p[1]*xe))), '--', color ='blue', label ="optimized data") 
plt.legend() 
plt.show() 
```

```python
from sklearn.metrics import r2_score
print('R^2: ', r2_score(ye, test_func(xe, p[0], p[1]))*100)
```
